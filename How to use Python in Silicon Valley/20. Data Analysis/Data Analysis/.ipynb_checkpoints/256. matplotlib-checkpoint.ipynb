```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('wansang')
plt.xlabel('kim')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'ro--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
