```python
# Machine Learning

# 30일 전 값들로 만들기
df_aapl['label'] = df_aapl['Close'].shift(-30)
df_aapl.tail(40)
```

```python
# Machine Learning

# Data with label line removed as X
# 라벨과 SMA(14일 이동평균)이 결과값이니까 X 값에서 제거해서 X로 만듬
X = np.array(df_aapl.drop(['label', 'SMA'], axis=1))
X.shape
```

```python
# When the featured value is inserted, the result becomes bad,
# so take the average and divide by the average difference to do scalin
# 값을 적절하게 변경(Preprocessing)
# 참고 읽기 자료 -> Data Preprocessing 핵심 전략(https://sacko.tistory.com/60)
X = sklearn.preprocessing.scale(X)
X.shape
```

```python
# Data from the past 30 days to be used for predicting

# 입력 데이터 -> X값
# 결과 데이터 -> X값에 30번째 전까지
# 예측 데이터 -> X값에 30번째 뒤부터

predict_data = X[-30:]
# Data inputted where past 30 days is removed
X = X[:-30]
y = np.array(df_aapl['label'])
# Correct label removing the past 30 days
y = y[:-30]
```

```python
# Divide into training data 80%, testing data 20%
# For the first function, data input, for the second function,  arranging correct label

# 결과데이터와 훈련데이터를 8:2 로 나눔
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(
    X, y, test_size = 0.2)
```

```python
# Learn from using training data

# linear-Regression으로 모델을 훈련함
lr = sklearn.linear_model.LinearRegression()
lr.fit(X_train,y_train)
```

```python
# 테스트 데이터의 점수를 계산

# Test using testing data
accuracy = lr.score(X_test, y_test)
accuracy
```

```python
# Predict

# 예상값 획득
predicted_data = lr.predict(predict_data)
predicted_data
```

```python
# 예상값 메모리 생성
df_aapl['Predict'] = np.nan

# 마지막 날은 마지막줄의 이름
last_date = df_aapl.iloc[-1].name

# 시간은 유닉스시간으로 표기
one_day = 86400
next_unix = last_date.timestamp() + one_day

# 예상값들을 하나씩 불러와서
for data in predicted_data:
    # 타임스테프에 다음날짜 생성
    next_date = datetime.datetime.fromtimestamp(next_unix)
    # 다음날짜에 증가
    next_unix += one_day
    # 다음날짜에 예상값 데이터 추가
    df_aapl.loc[next_date] = np.append([np.nan]* (len(df_aapl.columns)-1), data)

# 그래프 생성
df_aapl['Close'].plot(figsize=(15,6), color="green")
df_aapl['Predict'].plot(figsize=(15,6), color="orange")
plt.show()
```
